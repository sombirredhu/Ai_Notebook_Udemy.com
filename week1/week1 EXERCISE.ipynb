```python
# imports
import os
from openai import OpenAI
from dotenv import load_dotenv
from openai import Stream
```

```python
# constants
OLLAMA_BASE_URL = os.getenv("OLLAMA_BASE_URL")
OLLAMA_MODEL = os.getenv("OLLAMA_MODEL")
```

```python
# set up environment
client = OpenAI(base_url=OLLAMA_BASE_URL)
```

```python
# here is the question; type over this to ask something new

xsystemPrompt = """
Act as a think army tank, your role is to provide you though about country relations, also to provide info about how powerfull there army are, what should must do to win war against another community-contributions.
"""
question = """
Plase explain three countries relations with each other India, Russia and China
"""
```

```python
# Get gpt-4o-mini to answer, with streaming

try:
    from IPython.display import Markdown, display, update_display
    from IPython import get_ipython
except ImportError:
    Markdown = None
    display = None
    update_display = None
    get_ipython = None

def _running_in_jupyter_notebook() -> bool:
    """
    Check if running inside a Jupyter notebook.
    """
    if get_ipython is None:
        return False
    try:
        ip = get_ipython()
        return ip is not None and ip.__class__.__name__ == "ZMQInteractiveShell"
    except Exception:
        return False

stream = client.chat.completions.create(
    model=OLLAMA_MODEL,
    messages=[
        {"role": "system", "content": xsystemPrompt},
        {"role": "user", "content": question}
    ],
    stream=True
)
response = ""

# In notebook context, render Markdown streaming output
if display is not None and update_display is not None and Markdown is not None and _running_in_jupyter_notebook():
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        delta = chunk.choices[0].delta.content or ""
        response += delta
        if hasattr(display_handle, "update"):
            display_handle.update(Markdown(response))
        else:
            display_id = "answer_stream"
            display(Markdown(""), display_id=display_id)
            update_display(Markdown(response), display_id=display_id)
else:
    for chunk in stream:
        delta = chunk.choices[0].delta.content or ""
        response += delta
        print(delta, end="", flush=True)
    print()
```
